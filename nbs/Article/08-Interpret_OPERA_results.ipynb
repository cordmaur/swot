# Experiments Results

## Imports

```python
%load_ext autoreload
%autoreload 2

import pandas as pd
from tqdm.auto import tqdm

from swot_toolkit.analysis import open_sites_and_dates
from swot_toolkit.pipe2 import open_output_dir
```

## Read results

```python
sites_dates = open_sites_and_dates("/data/swot/output")
```

```python
results = pd.DataFrame()


for region, dates in tqdm(sites_dates.items(), desc="Regions"):
    for date in dates:
        base_dir, aoi, _ = open_output_dir(region=region, date=date)

        metrics_df = pd.read_parquet(base_dir / f"results/{region}_{date}_OPERA_metrics.parquet")
        metrics_df["Region"] = region
        metrics_df["Date"] = pd.to_datetime(date)
        results = pd.concat([results, metrics_df], axis=0)

columns = results.columns.to_list()[:-2]
```

```python
results = results.reset_index(names=["Metric"]).set_index(["Region", "Date", "Metric"])
results
```

## Calculate Overall Median

```python
# To calculate median over all regions and dates, we group by Metric
overall_results = results.groupby(level=["Metric"]).median().sort_index().round(3)
overall_results
```

## Calculate the Median by Region

```python
# To calculate median over all regions and dates, we group by Metric
region_results = results.groupby(level=["Region", "Metric"]).median().sort_index().round(3)
region_results
```

```python
# Prepare data
plot_data = region_results.reset_index().pivot(
    index="Region", columns="Metric", values=["opera_s2", "opera_s1"]
)
plot_data
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Rename regions for display
rename_regions = {
    "Rio_Madeira": "Rio Madeira",
    "Curua-Una": "Curuá-Una",
    "Rio_Negro": "Rio Negro",
    "Rio_Branco": "Rio Branco",
}

# Sort by OPERA S2 F1 score
plot_data = plot_data.sort_values(("opera_s2", "f1"), ascending=False)
plot_data.index = plot_data.index.map(lambda x: rename_regions.get(x, x))

# Define metrics and colors
metrics = ["f1", "iou", "precision", "recall"]
metric_labels = ["F1 Score", "IoU", "Precision", "Recall"]
colors = {
    "opera_s2": "#0077BB",  # Blue for S2
    "opera_s1": "#EE7733",  # Orange for S1
}

# Create subplots (2x2 grid)
fig = make_subplots(
    rows=2,
    cols=2,
    subplot_titles=metric_labels,
    vertical_spacing=0.22,
    horizontal_spacing=0.12,
)

# Add traces for each metric
for idx, (metric, label) in enumerate(zip(metrics, metric_labels, strict=True)):
    row = idx // 2 + 1
    col = idx % 2 + 1

    # OPERA S2
    fig.add_trace(
        go.Bar(
            name="OPERA S2" if idx == 0 else None,
            x=plot_data.index,
            y=plot_data[("opera_s2", metric)],
            marker_color=colors["opera_s2"],
            legendgroup="opera_s2",
            showlegend=(idx == 0),
            width=0.3,
            offsetgroup=0,
        ),
        row=row,
        col=col,
    )

    # OPERA S1
    fig.add_trace(
        go.Bar(
            name="OPERA S1" if idx == 0 else None,
            x=plot_data.index,
            y=plot_data[("opera_s1", metric)],
            marker_color=colors["opera_s1"],
            legendgroup="opera_s1",
            showlegend=(idx == 0),
            width=0.3,
            offsetgroup=1,
        ),
        row=row,
        col=col,
    )

# Update layout
fig.update_layout(
    barmode="group",
    plot_bgcolor="white",
    paper_bgcolor="white",
    font={"family": "Arial, sans-serif", "size": 16, "color": "black"},
    legend={
        "title": {"text": "Sensor", "font": {"size": 18}},
        "orientation": "h",
        "yanchor": "top",
        "y": -0.20,
        "xanchor": "center",
        "x": 0.5,
        "bgcolor": "rgba(255, 255, 255, 0.9)",
        "bordercolor": "black",
        "borderwidth": 1,
        "font": {"size": 16},
    },
    width=900,
    height=700,
    margin={"t": 80, "b": 120, "l": 80, "r": 40},
)

# Update subplot titles
for annotation in fig["layout"]["annotations"][:4]:
    annotation["font"] = {"size": 18, "color": "black"}

# Update all x-axes
for i in range(1, 5):
    fig.update_xaxes(
        showgrid=False,
        showline=True,
        linewidth=2,
        linecolor="black",
        mirror=True,
        ticks="outside",
        tickwidth=1.5,
        tickcolor="black",
        tickangle=-45,
        tickfont={"size": 14},
        row=(i - 1) // 2 + 1,
        col=(i - 1) % 2 + 1,
    )

# Update all y-axes
for i in range(1, 5):
    fig.update_yaxes(
        range=[0, 1.05],
        showgrid=False,
        showline=True,
        linewidth=2,
        linecolor="black",
        mirror=True,
        ticks="outside",
        tickwidth=1.5,
        tickcolor="black",
        tickmode="linear",
        tick0=0.0,
        dtick=0.2,
        tickfont={"size": 14},
        row=(i - 1) // 2 + 1,
        col=(i - 1) % 2 + 1,
    )

# Add y-axis labels to left plots only
fig.update_yaxes(title_text="Score", row=1, col=1, title_font={"size": 18})
fig.update_yaxes(title_text="Score", row=2, col=1, title_font={"size": 18})

fig.show()
```
