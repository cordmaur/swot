## Cross-track mean

```python
import numpy as np
import xarray as xr
```

```python
REGION = "Northeast"
DATE = sites_dates[REGION][0]
```

```python
swot, patches, no_data_masks = create_swot_mosaic(
    region_name=REGION,
    ref_date=DATE,
    exclude_flags=None,
    dst_crs=None,
    exclude_no_data=True,
    variable="cross_track",
)
```

```python
xr.concat([np.abs(patch["cross_track"]) for patch in patches], dim="idx").mean(dim="idx").mean()
```

```python
iou_results = results.reset_index().query("metric == 'iou'").set_index("level_0")
iou_results = iou_results.drop(columns=["metric"])["exclude no data"]
iou_results.index.name = "site"
iou_results
```

```python
results_dict = {}
for site, dates in sites_dates.items():
    date = dates[0]

    swot, patches, no_data_masks = create_swot_mosaic(
        region_name=site,
        ref_date=date,
        exclude_flags=None,
        dst_crs=None,
        exclude_no_data=True,
        variable="cross_track",
    )

    patches_abs = [np.abs(p["cross_track"].where(p["cross_track"] != 0)) for p in patches]

    patches_cube = xr.concat(patches_abs, dim="idx")
    results_dict[site] = {
        "cross_track_mean": patches_cube.mean(dim="idx").mean().item(),
        "iou": iou_results.loc[f"{site} {date}"].item(),
    }

# Prepare the graph
fig = go.Figure()

scatter = go.Scatter(
    x=[results_dict[site]["cross_track_mean"] for site in sites_dates],
    y=[results_dict[site]["iou"] for site in sites_dates],
    mode="markers+text",
    text=list(sites_dates.keys()),
    textposition="top center",
    marker={"size": 10, "color": "blue"},
)

fig.update_layout(
    xaxis={"title": "Mean Absolute Cross-Track Distance (m)"},
    yaxis={"title": "IoU Score"},
    title={"text": "IoU vs. Mean Absolute Cross-Track Distance", "x": 0.5},
)

fig.add_trace(scatter)
```

## Compute Results WITH dark water

```python
from swot_toolkit.metrics import calc_metrics, process_swot_mask
from swot_toolkit.pipe4 import calc_swot_metrics, open_datasets, quality_flags_bad
```

```python
region = "Rio_Branco"
date = sites_dates[region][0]
```

```python
metrics = ["iou", "f1", "precision", "recall"]
swot_metrics = calc_swot_metrics(region, date, metrics)
swot_metrics
```

```python
datasets = open_datasets(region, date)
```

```python
swot, patches, no_data_masks = create_swot_mosaic(
    region_name=region,
    ref_date=date,
    exclude_flags=["dark_water_suspect"],
    dst_crs=datasets["ref_mask"].rio.crs,
    exclude_no_data=True,
    variable="water_frac",
)
```

```python
swot_mask = process_swot_mask(swot, water_threshold=0.6)
dark_water = (swot_mask == 3).copy()
swot_mask = swot_mask.where(~dark_water, other=1)
dark_water_metric = calc_metrics(datasets["ref_mask"], swot_mask, metrics=metrics, binary=True)
dark_water_metric
```

```python
pd.concat(
    [
        swot_metrics[["exclude no data"]].rename(columns={"exclude no data": "SWOT no data"}),
        dark_water_metric.rename(columns={0: "Incl. Dark Water"}),
    ],
    axis=1,
)
```

```python
swot, patches, no_data_masks = create_swot_mosaic(
    region_name=region,
    ref_date=date,
    exclude_flags=quality_flags_bad,
    dst_crs=datasets["ref_mask"].rio.crs,
    exclude_no_data=False,
    variable="water_frac",
)
```

```python
swot_mask = process_swot_mask(swot, water_threshold=0.6)
swot_mask = swot_mask.where(~dark_water, other=1)
calc_metrics(datasets["ref_mask"], swot_mask, metrics=metrics, binary=True)
```
